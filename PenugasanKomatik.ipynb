```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

url ="https://raw.githubusercontent.com/RidwendDev/Magang-AI-Komatik-2023/main/penguins.csv"
dataset = pd.read_csv(url)

dataset.head()
```

```python
dataset[dataset['species'].apply(lambda x: 'Adelie' in x)]
```

```python
adelie = dataset[dataset['species'] == 'Adelie']
chinstrap = dataset[dataset['species'] == 'Chinstrap']
gentoo = dataset[dataset['species'] == 'Gentoo']

adelie.head()
```

```python
adelie['island'][adelie['sex']=='MALE'].value_counts()
```

```python
dataset[dataset['body_mass_g']==dataset['body_mass_g'].max()]
```

```python
dataset[dataset['body_mass_g']==dataset['body_mass_g'].min()]
```

```python
plt.rcParams['figure.figsize'] = (10,8)
```

```python
sns.histplot(adelie[adelie['sex'] == 'MALE']['body_mass_g'], label = 'Berat Jantan', color = 'red')
sns.histplot(adelie[adelie['sex'] == 'FEMALE']['body_mass_g'], label = 'Berat Betina')
plt.ylabel('Kepadatan')
plt.xlabel('Berat Badan')
plt.title('Persebaran Berat Badan Jantan dan Betina Spesies Adelie', size = 14)
plt.legend()
plt.show()
```

```python
sns.countplot(x='island',data=dataset, palette='hls')
```

```python
cowo = dataset['island'][dataset['sex']=='MALE'].value_counts()
plt.bar(['Torgersen','Biscoe','Dream'], cowo, color=['red', 'blue','yellow'])
plt.xlabel('Island')
plt.ylabel('Male')
plt.title('Persebaran Pinguins Jantan pada Setiap Pulau')
plt.show()
```

```python
sns.histplot(dataset[dataset['species'] == 'Adelie']['flipper_length_mm'], label = 'Sirip Adelie')
sns.histplot(dataset[dataset['species'] == 'Chinstrap']['flipper_length_mm'], label = 'Sirip Chinstrap', color = 'red')
sns.histplot(dataset[dataset['species'] == 'Gentoo']['flipper_length_mm'], label = 'Sirip Gentoo', color = 'green')

plt.ylabel('Kepadatan')
plt.xlabel('Panjang Sirip')
plt.title('Persebaran Panjang Sirip Setiap Spesies Pinguins', size = 14)
plt.legend()
plt.show()
```

```python
sns.boxplot(x=dataset['flipper_length_mm'], y=dataset['species'], orient='hls')
plt.ylabel('Species')
plt.xlabel('Panjang Sirip')
plt.title('Persebaran Panjang Sirip Pinguins Menurut Spesies Menggunakan Boxplot\n')
plt.show()
```
